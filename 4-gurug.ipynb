```python
plt.plot(df.age, df.charges)
plt.xlabel("yosh")
plt.ylabel("sugurta summasi")
plt.title("sugurta summasini yoshga nisbatan grafigi")
plt.show()
```

```python
df_ayol = df[df["sex"] =="female"]
df_erkak = df[df["sex"] =="male"]
```

```python
df_ayol.head()
```

```python
df_erkak.head()
```

```python
plt.plot(df_ayol.age, df_ayol.charges)
plt.plot(df_erkak.age, df_erkak.charges)
plt.xlabel("yosh")
plt.ylabel("sugurta summasi")
plt.title("sugurta summasini yoshga nisbatan grafigi")
plt.show()
```

```python
df_ayol = df_ayol[(df_ayol["age"] >= 20) & (df_ayol["age"] <= 30) ]
df_erkak = df_erkak[(df_erkak["age"] >= 20) & (df_erkak["age"] <= 30) ]
```

```python
plt.bar(df_ayol.age, df_ayol.charges)
plt.bar(df_erkak.age, df_erkak.charges)
plt.xlabel("yosh")
plt.ylabel("sugurta summasi")
plt.title("sugurta summasini yoshga nisbatan grafigi")
plt.show()
```

```python
df_ayol_far = df_ayol[(df_ayol["children"] >= 0)]
df_ayol_yuq = df_ayol[(df_ayol["children"] == 0)]
```

```python
plt.bar(df_ayol_far.age, df_ayol_far.charges)
plt.bar(df_ayol_yuq.age, df_ayol_yuq.charges)
plt.xlabel("yosh")
plt.ylabel("sugurta summasi")
plt.title("sugurta summasini yoshga nisbatan grafigi")
plt.show()
```

```python
df_e_far = df_erkak[(df_erkak["children"] >= 0)]
df_e_yuq = df_erkak[(df_erkak["children"] == 0)]
```

```python
plt.bar(df_e_far.age, df_e_far.charges)
plt.bar(df_e_yuq.age, df_e_yuq.charges)
plt.xlabel("yosh")
plt.ylabel("sugurta summasi")
plt.title("sugurta summasini yoshga nisbatan grafigi")
plt.show()
```
