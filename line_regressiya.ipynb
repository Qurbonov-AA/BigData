```python
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("insurance.csv")
```

```python
df.head()
```

```python
def show_charged(df):
    plt.plot(df.age, df.charges, 'go')
    plt.xlabel("age")
    plt.ylabel("charges")
    plt.title("yoshga nisbatan sug'urta")
    plt.show()
```

```python
show_charged(df)
```

```python
df_new = df[(df.children == 0) & (df.charges < 15000) & (df.sex == "female")  ]
df_new.shape
```

```python
df_new.loc[(df_new["age"] >= 0) & (df_new["age"] <= 30) & (df_new["charges"] > 6000), "charges"] = 3000
```

```python
df_new[(df_new["age"] >= 20) & (df_new["age"] <= 30) & (df_new["charges"] > 6000)]
```

```python
show_charged(df_new)
```

```python
train, test = train_test_split(df_new, test_size = 0.3, random_state=42)
train.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(train[["age"]], train.charges)
```

```python
from sklearn.metrics import mean_absolute_error
pred = lr.predict(test[['age']])
```

```python
mean_absolute_error(test.charges, pred)
```
