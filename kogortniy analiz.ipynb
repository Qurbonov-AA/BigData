```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime as dt
```

```python
df = pd.read_excel("Когортный анализ.xlsx")
df.head()
```

```python
#uznayem kolichestvo strok i kolumb
df.shape
```

```python
#proverim na polnatu i tipi kolumb
df.info()
```

```python
def get_month(x):
    return dt.datetime(x.year, x.month, 1)
```

```python
df["month_year"] = df['date'].apply(get_month)
```

```python
df.head(50)
```

```python
df['cohortMonth'] = df.groupby('user_id')['month_year'].transform('min')
```

```python
df.head(50)
```

```python
def get_date(df, columns):
    return df[columns].dt.year, df[columns].dt.month, df[columns].dt.day
```

```python
date_year, date_month, _ = get_date(df, 'date')
cohort_year, cohort_month, _ = get_date(df, 'cohortMonth')
```

```python
df['cohortIndex'] = (date_year-cohort_year)*12 + date_month-cohort_month + 1
```

```python
df.head()
```

```python
df.cohortIndex.unique()
```

```python
df_cohort = df.pivot_table(index='cohortMonth', 
                          columns='cohortIndex',
                          values = 'user_id',
                          aggfunc=len)
df_cohort
```

```python
df_cohort_price = df.pivot_table(index='cohortMonth', 
                          columns='cohortIndex',
                          values = 'price',
                          aggfunc='sum')
df_cohort_price
```

```python
first_column = df_cohort.iloc[:,0]
first_column_price = df_cohort_price.iloc[:,0]
```

```python
retension_cohort = df_cohort.div(first_column, axis=0)
retension_cohort.round(2)
retension_price = df_cohort_price.div(first_column_price, axis=0)
retension_price.round(2)
```

```python
def print_graph(df, name):
    plt.figure(figsize=(16,12))
    plt.title(f"kogortniy analiz {name}")
    sns.heatmap(data=df,
                annot = True,
                fmt = '.0%',
                vmin = 0.0,
                vmax = 0.5,
                cmap = 'YlGnBu'
    )
    plt.show()
```

```python
print_graph(retension_cohort, 'po user_id')
```

```python
print_graph(retension_price, 'po price')
```
